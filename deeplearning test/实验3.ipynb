```python
ad = pd.read_csv('Advertising.csv')
ad=(ad - ad.min()) / (ad.max() - ad.min())
```

```python
feature_cols = ['TV', 'Radio', 'Newspaper']
X = ad[feature_cols]
X = ad[['TV', 'Radio', 'Newspaper']]
y = ad.Sales
y = ad['Sales']
```

```python
X_train,X_test, y_train, y_test = (X[:160],X[161:], y[:160],y[161:])
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
fig=plt.figure(figsize=(16,10))
for i in range(9):
    poly = PolynomialFeatures(degree=i+1,interaction_only=False,include_bias=True)
    x_train_poly_s = poly.fit_transform(X_train)
    x_test_poly_s = poly.fit_transform(X_test)
    
    linreg = LinearRegression()
    linreg.fit(x_train_poly_s,y_train)
    y_pred = linreg.predict(x_test_poly_s)
    
    plt.subplot(3,3,i+1)
    plt.plot(range(len(y_pred)),y_pred,'g',label="predict")
    plt.plot(range(len(y_pred)),y_test,'r',label="test")
    plt.legend(loc="upper right") #显示图中的标签
    plt.xlabel("the number of sales")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2,interaction_only=True,include_bias=True)
x_train_poly_s = poly.fit_transform(X_train)
x_test_poly_s = poly.transform(X_test)



linreg = LinearRegression()
linreg.fit(x_train_poly_s,y_train)
y_pred = linreg.predict(x_test_poly_s)
    

plt.plot(range(len(y_pred)),y_pred,'g',label="predict")
plt.plot(range(len(y_pred)),y_test,'r',label="test")
plt.legend(loc="upper right") #显示图中的标签
plt.xlabel("the number of sales")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

from sklearn.model_selection import train_test_split

X = iris.data[:, 1:4]       
Y = iris.target

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
for i in range(9):
    po = PolynomialFeatures(degree=i+1,interaction_only=False,include_bias=True)
    x_train_po = po.fit_transform(x_train)
    x_test_po = po.fit_transform(x_test)


    from sklearn.linear_model import LogisticRegression

    lr = LogisticRegression(penalty='l2',solver='newton-cg',multi_class='multinomial')
    lr.fit( x_train_po,y_train)
    print("-----------项数为"+str(i+1)+"-----------")
    print("Logistic Regression模型训练集的准确率：%.3f" %lr.score(x_train_po, y_train))
    print("Logistic Regression模型测试集的准确率：%.3f" %lr.score(x_test_po, y_test))
```
